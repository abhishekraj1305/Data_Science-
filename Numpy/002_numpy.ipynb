```python
#vector

vector = np.array([35,8,78,90])
print("Vector ", vector)

matrix = np.array([[35,8,78,90],[56,78,33,74]])
print("matrix ", matrix)

tensor = np.array([[35,8,78,90],[56,78,33,74],[86,70,93,34],[37,85,78,80]])
print("tensor ", tensor)
```

```python
arr = np.array([[35,8,78,90],[56,78,33,74],[86,70,93,34],[37,85,78,80]])        
print("arr ", arr)

print("shape", arr.shape)
print("Dimension", arr.ndim)
print("size", arr.size)
print("dtype", arr.dtype)
print("astype", arr.astype)
print("congugate", arr.conjugate)


```

```python
arr = np.arange(39)
print("original",arr)

resh = arr.reshape((3,13))
print("reshape", resh)

flattened = resh.flatten()
print('flatten reshaped',flattened)

raveld = resh.ravel()
print('Raveled: ', raveld)


transpose = resh.T
print("transposed array ", transpose)
```
