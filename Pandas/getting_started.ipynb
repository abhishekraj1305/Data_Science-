###creating dataframe


```python
import pandas as pd

# List of lists
data = [
    [1, "Abhishek", 25],
    [2, "Rahul", 27],
    [3, "Ankit", 26]
]

df1 = pd.DataFrame(data, columns=["ID", "Name", "Age"])

print("DataFrame from List:")
print(df1)
```

```python
import pandas as pd

data = {
    "ID": [1,2,3],
    "Name": ["Abhishek","Rahul","Ankit"],
    "Age": [25,27,26]
}

df2 = pd.DataFrame(data)

print("\nDataFrame from Dictionary:")
print(df2)
```

```python
import pandas as pd

s1 = pd.Series([1,2,3])
s2 = pd.Series(["Abhishek","Rahul","Ankit"])
s3 = pd.Series([25,27,26])

df3 = pd.DataFrame({
    "ID": s1,
    "Name": s2,
    "Age": s3
})

print("\nDataFrame from Series:")
print(df3)
```

```python
import pandas as pd

data = [
    {"ID":1, "Name":"Abhishek", "Age":25},
    {"ID":2, "Name":"Rahul", "Age":27},
    {"ID":3, "Name":"Ankit", "Age":26}
]

df4 = pd.DataFrame(data)

print("\nDataFrame from List of Dictionary:")
print(df4)
```

```python
import pandas as pd

df5 = pd.DataFrame()

print("\nEmpty DataFrame:")
print(df5)
```

```python
import pandas as pd

data = {
    "Name": ["Abhishek","Rahul","Ankit"],
    "Age": [25,27,26]
}

df6 = pd.DataFrame(data, index=["A","B","C"])

print("\nDataFrame with Custom Index:")
print(df6)
```

```python
import pandas as pd
import numpy as np

data = np.array([
    [1,25],
    [2,27],
    [3,26]
])

df7 = pd.DataFrame(data, columns=["ID","Age"])

print("\nDataFrame from NumPy:")
print(df7)
```

```python
import pandas as pd

df8 = pd.DataFrame()

df8["ID"] = [1,2,3]
df8["Name"] = ["Abhishek","Rahul","Ankit"]
df8["Age"] = [25,27,26]

print("\nDataFrame Column Wise:")
print(df8)
```
